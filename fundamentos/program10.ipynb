# Estruturas de Dados - Listas

```python
# Lista são coleções ordenadas e imutaveis de itens. Podem conter diferentes tipos de dados
frutas = ["maçã", "banana", "laranja", "abacaxi"]
print(frutas)
```

```python
type(frutas)
```

```python
# Acessando um item pelo indice
print(f"A primeira fruta é: {frutas[0]}")
print(f"A ultima fruta é: {frutas[-1]}") #vai ate o fim da lista com indice -1
```

```python
# Adicionando um item ao final da lista
frutas.append("uva")
print(f"Lista após adicionar 'uva': {frutas}")
```

```python
# Remover um item
frutas.remove("laranja")
print(f"Lista após remover 'laranja': {frutas}")
```

```python
# Modificando um item
frutas[0] = "morango"
print(f"Lista após modificar 'morango': {frutas}")
```

```python
#imprimindo diretamente
print(f"A lista tem {len(frutas)} frutas.")
```

```python
#Deletando a lista
del frutas
```

```python
# Lista sem acesso
print(frutas)
```
