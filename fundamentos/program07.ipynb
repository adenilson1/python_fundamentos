## Operadores de Comparação

```python
# Definindo variáveis
x = 5
y = 18
```

```python
# Operador "maior que "
x > y
```

```python
# Operador "menor que "
x < y
```

```python
# Operador "igual a "
x == y
```

```python
# Operador "diferente de "
x != y
```

```python
# Operador "maior ou igual "
x >= 5
```

```python
# Operador "menor ou igual "
x <= y
```
