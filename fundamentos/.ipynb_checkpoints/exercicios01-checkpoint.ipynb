#### Exercício 2

Crie duas variáveis, numero1 e numero2, e atribua a elas valores inteiros. Calcule a soma, subtração, multiplicação e divisão dessas variáveis e imprima os resultados.

```python
#Entrada de dados
print(f"Digite dois numeros inteiros: ")
number_first = input("Primeiro numero: ")
number_first_int = int(number_first)
number_second = input("Segundo numero: ")
number_second_int = int(number_second)

#saida de dados
print(f"A soma de {number_first} + {number_second} = {number_first_int + number_second_int}")
print(f"A subtração de {number_first} - {number_second} = {number_first_int - number_second_int}")
print(f"A multiplicação de {number_first} * {number_second} = {number_first_int * number_second_int}")
print(f"A divisão de {number_first} / {number_second} = {(number_first_int / number_second_int):.2f}")
```

#### Exercício 8

Crie duas variáveis booleanas: 

- chovendo = True
- guarda_chuva = False

Use operadores lógicos para verificar se uma pessoa vai se molhar (se está chovendo E ela não tem guarda-chuva).

```python
chovendo = True
guarda_chuva = False
print("A pessoa vai se molhar?")
if(chovendo and guarda_chuva) == True:
    print("NÃO")
elif(not(chovendo) and guarda_chuva) == True:
    print("NÃO")
elif(not(chovendo) and not(guarda_chuva)) == True:
    print("NÃO")
else:
    print("SIM")
```

Crie dois conjuntos: 

- set_a = {1, 2, 3, 4}

Encontre e imprima a interseção entre os dois conjuntos (os elementos que estão em ambos).

```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}
```

```python
conjunto_intersecao = set_a.intersection(set_b)
```

#### Exercício 30

Escreva um programa que peça ao usuário para digitar sua altura em metros (ex: 1.75) e seu peso em quilogramas (ex: 68.5). Calcule o Índice de Massa Corporal (IMC) usando a fórmula IMC = peso / (altura * altura) e imprima o resultado formatado com duas casas decimais.

```python
print("Calculo IMC")
altura = float(input("Digite a altura: "))
peso = float(input("Digite o peso: "))
imc = peso/(altura**2)
print(f"\nCom a {altura}m e o peso {peso}Kg: IMC = {imc:.2f}")
```
