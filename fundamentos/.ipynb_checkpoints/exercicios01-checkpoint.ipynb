#### Exercício 2

Crie duas variáveis, numero1 e numero2, e atribua a elas valores inteiros. Calcule a soma, subtração, multiplicação e divisão dessas variáveis e imprima os resultados.

```python
#Entrada de dados
print(f"Digite dois numeros inteiros: ")
number_first = input("Primeiro numero: ")
number_first_int = int(number_first)
number_second = input("Segundo numero: ")
number_second_int = int(number_second)

#saida de dados
print(f"A soma de {number_first} + {number_second} = {number_first_int + number_second_int}")
print(f"A subtração de {number_first} - {number_second} = {number_first_int - number_second_int}")
print(f"A multiplicação de {number_first} * {number_second} = {number_first_int * number_second_int}")
print(f"A divisão de {number_first} / {number_second} = {(number_first_int / number_second_int):.2f}")
```

#### Exercício 3 

Qual é a principal diferença entre uma variável de escopo local e uma de escopo global?

Resposta: um variável so atua dentro de um bloco, escopo,  espercifico de um codigo, em quanto uma variavel global pode atua em qualquer 
bloco de codigo quando assim for requisitada.

#### Exercício 4

Crie uma variável chamada saldo com o valor 500.50 (float). Em seguida, crie uma variável saque com o valor 200.25 (float). Subtraia o saque do saldo e imprima o saldo final formatado para duas casas decimais.

```python
valor_saldo = 500.50
valor_saque = 200.25
valor_final = valor_saldo - valor_saque
print(f"Saldo final é R$ {valor_final:.2f}")
```

#### Exercício 6

Crie duas variáveis: 

- idade_ana = 25
- idade_beto = 30

Use operadores de comparação para verificar se a idade de Ana é menor que a de Beto e imprima o resultado booleano.

```python
#Variáveis
idade_ana = 25
idade_beto = 30

print(f"{idade_ana} é maior ou igual a {idade_beto}: {idade_ana >= idade_beto}")
print(f"{idade_ana} é menor ou igual a {idade_beto}: {idade_ana <= idade_beto}")
print(f"{idade_ana} é igual a {idade_beto}: {idade_ana == idade_beto}")
```

#### Exercício 7

Receba um número inteiro do usuário e use o operador de módulo (%) para verificar se o número é par ou ímpar. Imprima o resultado.

Obs: Não estudamos ainda os condicionais, mas pesquise se necessário e tente resolver.

```python
print("O número é par ou impar")
number = float(input("Digite um número: "))
if(number%2 == 0):
    print("PAR")
else:
    print("ÍMPAR")
```
