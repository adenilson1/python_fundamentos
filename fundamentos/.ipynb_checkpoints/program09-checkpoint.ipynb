# Manipulação de Strings

```python
# Strings são sequenciias de caracteres e possuem muitos métodos 
```

```python
# Concatenação
nome = "Maria"
saudacao = "Olá, " + nome + "!"
print(saudacao)
```

```python
#Tamanho da string
#frase
frase = "  Aprender Python é muito divertido!  "
print(f"Tamanho da frase: {len(frase)}")
```

```python
#Maiúsculas e minúsculas
print(f"Maiúsculas: {frase.upper()}")
print(f"Minúscula: {frase.lower()}")
```

```python
#Remover espaços em branco do inicio e do fim
frase_sem_espacos = frase.strip()
print(f"Frase sem espaços: '{frase_sem_espacos}'")
```

```python
#Substituir text
print(f"Substituir 'divertido' por 'legal': {frase_sem_espacos.replace('divertido', 'legal')}")
```

```python
#Fatiando (slinig) - Acessando partes de uma string
#o indice em Python em 0
print(frase_sem_espacos)
print(f"O primeiro caracter: {frase_sem_espacos[0]}")
print(f"O primeiro caracter: {frase_sem_espacos[9:15]}") # do indice 9 ao 14
```
