# Conversão Entre Tipos de Dados - type Casting

```python
#Conversão de string para número integer
numero_em_texto = "123"
numero_inteiro = int(numero_em_texto)
print(f"String '{numero_em_texto}' para Inteiro: {numero_inteiro}, Tipo: {type(numero_inteiro)}")
```

```python
#Conversão de string para número float
numero_em_texto = "123"
numero_float = float(numero_em_texto)
print(f"String '{numero_em_texto}' para Float: {numero_float}, Tipo: {type(numero_float)}")
```

```python
#Conversão de string para número integer
idade = 23
idade_texto = str(idade)
print(f"Inteiro {idade} para String: '{idade_texto}', Tipo: {type(idade_texto)}")
```

```python
#Conversão entre estruturas de dados
lista_com_duplicada = [1,2,2,3,4,4,4,5]
conjunto_unico = set(lista_com_duplicada)
lista_sem_duplicada = list(conjunto_unico)
```

```python
print(f"\nLista original: {lista_com_duplicada}")
print(f"\nConversão para Conjunto (remove duplicadas): {conjunto_unico}")
print(f"\nConvertida de volta pra Lista: {lista_sem_duplicada}\n")
```
