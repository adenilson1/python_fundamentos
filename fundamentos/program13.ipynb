# Estruturas de Dados - Conjuntos (Set)

```python
# Conjuntos: São coleções ná ordenadas de itens únicos e mutáveis. Eles são úteis para remover duplicadas 
# e realizar operações matemáticas de conjuntos (união, interseção)
```

```python
#Criando um conjunto 
numeros = {1, 2 , 3, 4, 2, 3, 5}
print(f"Conjuntos de números (sem duplicadas): {numeros}")
```

```python
type(numeros)
```

```python
# Adicionando um item
numeros.add(6)
print(f"Após adicionar o valor 6: {numeros}")
```

```python
#Removendo um item
numeros.remove(2)
print(f"Após remover o 2: {numeros}")
```

```python
#Operação de conjunto
conjunto_a = {1,2,3,4}
conjunto_b = {3,4,5,6}
```

```python
# União 
uniao = conjunto_a.union(conjunto_b)
print(f"União de A e B: {uniao}")
```

```python
#Interseção
intersecao = conjunto_a.intersection(conjunto_b)
print(f"Interseção de A e B: {intersecao}")
```
