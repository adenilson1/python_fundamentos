# Estruturas de Dados - Tuplas

```python
# Tuplas: São coleções ordenadas e imutaveis de itens. uma vez criada, não pode ser alteradas
```

```python
#Criando uma tupla
coordenadas = (10.0, 20.5)
print(f"Tupla de coordenadas : {coordenadas}")
```

```python
#Acessando um item pelo indice
print(f"Coordenada X: {coordenadas[0]}")
print(f"Coordenada Y: {coordenadas[1]}")
```

```python
#Tentativa de modificar um item de um tupla resultará em erro
coordenadas[0] = 15.0 
```

```python
#Tuplas são uteis para dados que não podem ser alterados, como meses do ano, coordenadas, etc
dias_da_semana = ("Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sabádo", "Domingo")
print(f"O primeiro dia da semana é: {dias_da_semana[0]}")
```
