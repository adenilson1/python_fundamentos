# Escopo de variáveis em código python

```python
# Variável global
saudacao = "Olá Mundos !!!"
```

```python
nome = "Aluno"

#Funçaõ
def minha_funcao():
    nome = "Ana"
    print(f"\nDentro da função: {nome}")
    print(f"\nAcessando a variável global de dentro da funcao: {saudacao}")

minha_funcao()

print(f"\nFora da função: {saudacao}")
print(f"\nFora da função: {nome}")
```
