## Comprehesions(List, Set, Dict e Generator) - 1/2

#### Estas estruturas são considerados construtores sintáticos( syntactic constructs) ou mais formalmente, expressões de compreensão (comprehension expressions).

#### Na documentação oficial:
#### .List comprehension -> expressão que gera listas
#### .Set comprehension -> expressão que gera Conjuntos
#### .Dict comprehension -> expressão que gera Diciónarios
#### .Generator expressions -> expressão que gera iteradores (e pode ser convertido em tupla, lista, etc.).

#### Ou seja, o termo mais amplo é comprehension: uma forma mais curta e expressiva de construir coleções (ou genedores) a partir de iteráveis com filtros e transormações aplicadas inline

```python
# Criando uma lista de quadrados dos números de 0 a 9
quadrados = [x ** 2 for x in range(10)]
print(f"\nQuadrados de 0 a 9: {quadrados}")
             
```

```python
#Criando um alista de numeros pares de 0 a 20
pares = [x for x in range(21) if x % 2 == 0]

print(f"Números pares de 0 a 20: {pares}")
```

```python
# Criando um dicionario com numeros quadrados
quadrados_dcit = {x: x ** 2 for x in range(6)}
print(quadrados_dcit)
```
