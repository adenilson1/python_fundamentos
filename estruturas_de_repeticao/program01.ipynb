# Tomada de Decisão com Estruturas Condicionais

```python
nota = 8.5
```

```python
if(nota >= 7.0):
    print("Aprovado!")
else:
    print("Reprovado!")
```

```python
idade = 17
```

```python
if(idade < 18):
    print("Você é menor de idade.")
elif (idade >= 18 and idade < 65):
    print("Você é adulto.")
else:
    print("Você é idoso.")
```
