# Estrutura de Repetição - Loops com condicional 

```python
# Tuplas
numeros = (3, 7, 10, 15, 20)

for n in numeros:
    if n % 2 == 0:
        print(f"{n} é par")
```

```python
# Lista de nomes
nomes = ["Ana", "Bruno", "Carlos", "Amanda", "Beatriz"]

for nome in nomes:
    if nome.startswith("A"):
        print(f"Nome encontrado com 'A': {nome}")
```

```python
# Dicionário 
produtos = {"arroz": 25, "feijão": 12, "carne": 45, "marcarrão": 8}
for item, preco in produtos.items():
    if preco > 20:
        print(f"{item} custa R$ {preco} reais (acima de 20)")
```
