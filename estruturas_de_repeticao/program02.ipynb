# Estruturas de Repetição - Loop For - Parte 1/2

### Loop for: é usado para iterar sobre uma sequência (como lista, tupla, dicionário, conjunto ou string

```python
#Definição de lista
frutas = ["maçã", "banana", "cereja"]

#Imprimir mensagem
print("Frutas disponíveis:")

#loop pela lista
for fruta in frutas:
    print(f"- {fruta}")
```

```python
#Criando uma tuplas
cores = ("vermelho", "verde", "azul")

#loop
for cor in cores:
    print(cor)
```

```python
#Criando um dicionario com o numero de cursos
formacoes = {"Formação Cientista de Dados": 6, "Formação Analista de Dados": 4, "Formação Engenheiro de Dados": 5}

#loop
for chave, valor in formacoes.items():
    print(chave, ":", valor)
```

```python
#Exemplo com função range()
print("\nContagem até 5:")
for numero in range(6):
    print(numero
```
