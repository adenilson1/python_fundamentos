# Funções

Funções são blocos de códigos reutilizáveis que realizam uma tarefa específica

```python
# Definição de uma função simple
def saudacao():
    """Está função exibe uma saudação simple."""
    print("\nOlá! Bem-vindo ao Python.")
```

```python
#Chamando a função
saudacao()
```

```python
# Definição de uma função com parametros
def soma_numeros(a,b):
    return a + b
```

```python
# Chamando a função e armazenando o resultado em uma variavel
resultado = soma_numeros(5,3)

print(f"O resultado da soma é: {resultado}")
```
