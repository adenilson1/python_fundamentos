## Comprehesions(List, Set, Dict e Generator) - 2/2

```python
# Criando um dicionario com numeros quadrados
quadrados_dict = {x : x ** 2 for x in range(6)}
print(quadrados_dict)
```

```python
type(quadrados_dict)
```

```python
# Conjunto de quadrados(sem repetir)
quadrados_set = {x ** 2 for x in [1, 2, 2, 3, 3, 4]}
print(quadrados_set)
```

```python
type(quadrados_set)
```

```python
#Gererator expression (não é tupla ainda)
gen = (x ** 2 for x in range(6))
print(gen)

# Convertendo em tupla
quadrados_tuple = tuple( x ** 2 for x in range(6))
print(quadrados_tuple)
```

```python
type(quadrados_tuple)
```

Um generator é um iterador que não armazena todos os dados de uma vez, mas sim gera cada valor sob demanda. No caso, gen não é uma lista de quadrados de 0 a 5, mas um objeto que sabe como calcular esses valores quando solicitado. A grande vanteagem do genereator é que ele economiza memoria
