# Estruturas de Repetição - Loop For - Parte 2/2

```python
# Criando uma estrutura de repetição com numeros de cursos
formacoes = {"Formação Cientista de Dados": 6, "Formação Analista de Dados": 4, "Formação Engenheiro de Dados": 5}

#Loop
for chave, valor in formacoes.items():
    print(chave, ":", valor)
```

```python
# Exemplo dom range()
print("\nContagem até 5")
for numero in range(6):
    print(numero)
```

## Loop While

#### O loop While executa um bloco de código enquanto uma condição for verdadeira

```python
contador = 5

print("Contagem regressiva")

while contador > 0:
    print(contador)
    contador -= 1
```
