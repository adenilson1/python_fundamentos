# Parâmetros e Argumentos de Funções

```python
#Argumentos posicionais
def apresentacao(nome, idade):
    print(f"Nome: {nome}, idade: {idade}")
```

```python
apresentacao("Ana", 25)
```

```python
#Chamando de forma errada
apresentacao(25, "Ana")
```

```python
# nomeando argumentos
apresentacao(idade = 30, nome = "Bob")
```

```python
#Paramentros com valores padrão (default)
def saudacao_completa(nome, saudacao = "Olá"):
    print(f"{saudacao}, {nome}")
```

```python
saudacao_completa()
```

```python
saudacao_completa("Maria")
```

```python
saudacao_completa("Bob", "Bom dia")
```
